```python
user_ratings=movies.pivot_table(index=['title'],columns=['user_id'],values='rating')
user_ratings.head()
```

```python
x=movies.groupby('user_id').count()['rating']>30
rated_person=x[x].index
```

```python
rated_person
```

```python
filtered_rating=movies[movies['user_id'].isin(rated_person)]
```

```python
y=filtered_rating.groupby('title').count()['rating']>=50
famous_movies=y[y].index
```

```python
final_rating=filtered_rating[filtered_rating['title'].isin(famous_movies)]
```

```python
final_rating.drop_duplicates()
```

```python
pt=final_rating.pivot_table(index='title',columns='user_id',values='rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt)
```

```python
cosine_similarity(pt).shape
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommned(title):
    index=np.where(pt.index==title)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[0:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```
